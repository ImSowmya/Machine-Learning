### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read CSV and analyse

```python
data = pd.read_csv("multiple_linear_regression.csv")
data.head()
```

```python
data.info()
```

```python
data.shape
```

### Split data

Input parameters

```python
input_columns = data.iloc[:,:-1]
input_columns
```

```python
output_column = data.loc[:,"Target"]
output_column
```

### Splitting dataset using sklearn

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input_columns, output_column, test_size=0.3, random_state=43)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

# MULTIPLE LINEAR REGRESSION

### TRAINING

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

### PREDICTION

```python
pred = lr.predict(x_test)
```

### COST

```python
from sklearn.metrics import mean_squared_error
cost = mean_squared_error(y_test, pred)
cost
```

```python
plt.plot(x_test,y_test,'*',color ='green')
plt.plot(x_test,pred,'+',color = 'red')
plt.title("Performance Testing")
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()
```
