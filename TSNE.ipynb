```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
features = list(data.columns[:-1])
target = data.columns[-1]
```

```python
target
```

```python
data_norm = data.copy()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
data_norm[features] = sc.fit_transform(data_norm[features])
```

TSNE

```python
from sklearn.manifold import TSNE

tsne = TSNE(learning_rate=450, n_components=2)
x_tsne = tsne.fit_transform(data_norm[features])
```
